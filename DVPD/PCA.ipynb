```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from sklearn.datasets import load_breast_cancer
cancer_dataset=load_breast_cancer()
```

```python
type(cancer_dataset)
```

```python
cancer_dataset.keys()
```

```python
cancer_dataset.DESCR
```

```python
df=pd.DataFrame(cancer_dataset['data'],columns=cancer_dataset['feature_names'])
df.head()
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
scaler.fit(df)
```

```python
# standardization
scaled_data=scaler.transform(df)
scaled_data #ndarray
scaled_data
```

```python
# Applying PCA Algorithm
from sklearn.decomposition import PCA
```

```python
pca=PCA(n_components=2)
```

```python
data_pca=pca.fit_transform(scaled_data)
data_pca
```

```python
pca.explained_variance_
```

```python
# Plot
plt.figure(figsize=(10,12))
plt.scatter(data_pca[:,0],data_pca[:,1],c=cancer_dataset['target'],cmap='plasma')
plt.xlabel("PC1")
plt.ylabel("PC2")
plt.show()
```

```python
# using seaborn
# Create a DataFrame to use with Seaborn
df = pd.DataFrame(data_pca, columns=['PCA1', 'PCA2'])
df['target'] = cancer_dataset['target']

# Plot with hue to color by target class
sns.scatterplot(data=df, x='PCA1', y='PCA2', hue='target', palette='plasma')
plt.xlabel('PCA Component 1')
plt.ylabel('PCA Component 2')
plt.show()
```
