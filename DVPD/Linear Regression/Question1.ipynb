```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df=pd.read_csv("data_13_11_1 - data_13_11_1.csv")
df.head()
```

```python
df.info()
```

```python
plt.figure(figsize=(5,5))
sns.scatterplot(x='X',y='Y',data=df)
plt.xlabel("X")
plt.ylabel('Y')
plt.show()
```

```python
x=df['X'].values
y=df['Y'].values
```

```python
def loss(y,y_pred):
    return ((y-y_pred)**2).mean()


def gradient_descent(x,y,m,c,learning_rate):
    N=len(y)
    y_pred=m*x+c
    dm=-(2/N)*sum(x*(y-y_pred))
    dc=-(2/N)*sum(y-y_pred)
    m=m-learning_rate*dm
    c=c-learning_rate*dc
    return m,c
```

```python
#Training model
m=0
c=0
iteration=1000
learning_rate=0.0001

loss_list=[]
for i in range(iteration):
    y_pred=m*x+c
    loss_list.append(loss(y,y_pred));
    m,c=gradient_descent(x,y,m,c,learning_rate)
    
```

```python
# ploting loss fxn
plt.figure(figsize=(10,7))
plt.plot(range(iteration),loss_list)
plt.xlabel("Iteration")
plt.ylabel("Loss")
plt.title("Loss Function")
plt.show()
```

```python
#Plotting best fit line
plt.scatter(x, y, color='blue', label='Data Points')
plt.plot(x, m * x + c, color='red', label='Best Fit Line')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Best Fit Line using Gradient Descent')
plt.legend()
plt.show()
```
